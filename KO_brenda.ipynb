```python
import requests

def get_enzyme_info(ko_number):
    url = f"http://rest.kegg.jp/get/{ko_number}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        return None

ko_number = "K00001"
enzyme_info = get_enzyme_info(ko_number)
if enzyme_info:
    print(enzyme_info)
else:
    print("No information found for the given KO number.")
```

```python
import requests

def get_brenda_info(ec_number):
    url = f"https://www.brenda-enzymes.org/enzyme.php?ecno={ec_number}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        return None

ec_number = "1.1.1.1"  # 替换为你要查询的 EC 号
brenda_info = get_brenda_info(ec_number)
if brenda_info:
    print(brenda_info)
else:
    print("No information found for the given EC number.")
```

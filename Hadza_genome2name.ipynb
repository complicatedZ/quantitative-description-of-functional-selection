```python
import numpy as np
import scipy.io as sio
import os
import pandas as pd

# Load the .mat files
data = sio.loadmat('Hadza_converted.mat')
data2  = pd.read_csv('SupplementalTable_S2.csv')
```

```python
genomes = np.array([gen_name[0] for gen_name in data['genomes'].flatten()])
usamples = np.array([sample_name[0] for sample_name in data['usamples'].flatten()])
relabus = data['relabus']
kegg_per_genome = data['keggs_per_genome']
```

```python
def find_indices(arr1, arr2):
    """
    Returns the indices of the elements in arr1 as they appear in arr2.

    Parameters:
    arr1 (array-like): Array whose elements' indices in arr2 are to be found.
    arr2 (array-like): Array from which the indices are to be found.

    Returns:
    array: Indices of elements in arr1 as they appear in arr2.
    """
    index_dict = {value: idx for idx, value in enumerate(arr2)}
    return np.array([index_dict[value] for value in arr1])
```

```python
genome_dataset = np.array(data2['genome']).astype(str)
```

```python
GTDB_taxonomy = np.array(data2['GTDB_r95_taxonomy'])
```

```python
corrinds = np.zeros(genomes.shape)
for i, ge_name in enumerate(genomes):
    try:
        corrinds[i] = np.where(genome_dataset==(ge_name.split('.')[0]))[0]
    except:
        corrinds[i] = -1
```

```python
corrinds.astype(int)
```

```python
corrinds = corrinds.astype(int)
genome2name = []
for i in range(len(genomes)):
    if corrinds[i] != -1:
        try:
            genome2name.append(GTDB_taxonomy[corrinds[i]].split(';'))
        except:
            genome2name.append('Not annotated')
    else:
        genome2name.append(['Not in dataset'])
```

```python
max_length = max(len(subseq) for subseq in genome2name)
genome2name_array = np.array([subseq + [None] * (max_length - len(subseq)) for subseq in genome2name])
```

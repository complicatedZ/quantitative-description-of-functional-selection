```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('genome2name.csv')
```

```python
EQO_selected = np.array(pd.read_csv('temp.csv'))[:,1]==1
percent_selected = np.array(pd.read_csv('temp.csv'))[:,2]
```

```python
vanish = np.zeros(len(percent_selected))
blossom = np.zeros(len(percent_selected))
for i, value in enumerate(percent_selected):
    if value<=100 and value >=95:
        blossom[i]=1
    elif value>=-100 and value <=-95:
        vanish[i]=1
```

```python
blossom
```

```python
group_counts = df['Family'].value_counts()
group_counts = group_counts[1:]
top_15 = group_counts.head(15)
others_count = group_counts[15:].sum()

# 将 "Others" 添加到前十五个种类中
top_15['Others'] = others_count
```

```python
plt.rcParams['font.family'] = 'Arial'
plt.figure(figsize=(8, 6))
plt.pie(top_15, labels=top_15.index, autopct=lambda p: f'{int(p * sum(top_15) / 100)}', startangle=140)
plt.title('Family distribution')
plt.show()
```

```python
group_counts = df['Family'][EQO_selected].value_counts()
group_counts = group_counts[1:]
top_15 = group_counts.head(15)
others_count = group_counts[15:].sum()

# 将 "Others" 添加到前十五个种类中
top_15['Others'] = others_count
```

```python
plt.rcParams['font.family'] = 'Arial'
plt.figure(figsize=(6, 10))
plt.pie(top_15, labels=top_15.index, autopct=lambda p: f'{int(p * sum(top_15) / 100)}', startangle=140)
plt.title('Family distribution(after selected by EQO)')
plt.show()
```
